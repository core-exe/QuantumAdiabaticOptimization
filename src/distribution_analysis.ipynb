```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('dark')
```

```python
path = "output/qubit_10_t_256/"
linear_dist = np.load(path + 'Linear_distribution.npy')
opt_dist = np.load(path + 'Opt_distribution.npy')
np.mean(linear_dist), np.mean(opt_dist)
```

```python
maximum = max(np.max(linear_dist), np.max(opt_dist))
interval = 0.01
bins = [i*interval for i in range(int(maximum/interval)+1)]
```

```python
fig = plt.figure(figsize=(8, 6))
plt.hist(linear_dist, bins = bins, alpha=0.7, label='linear', color='#7FC8F8')
plt.hist(opt_dist, bins = bins, alpha=0.7, label='best', color='#FFE45E')
plt.legend()
plt.yticks([])
plt.xlabel("success probability")
plt.title("3SAT, n=10, m=42, distribution of linear and best annealing path")
plt.savefig(path+'distribution.pdf')
```
